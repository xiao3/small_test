```python
import pandas as pd
import geopandas as gpd
```

```python
df_covid = pd.read_csv("./data/03-30-2020.csv")
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.head()
```

```python
world.plot(figsize=(20,10))
```

```python
df_covid.head()
```

```python
df_covid_sum = df_covid.groupby(['Country_Region'])['Confirmed'].agg('sum').reset_index()
df_covid_sum.sort_values('Confirmed',ascending=False).head(10)
```

```python
df_covid.groupby(['Country_Region'])['Confirmed','Deaths', 'Recovered'].sum().sort_values('Confirmed',ascending=False).head(10).plot(kind='bar',figsize=(20,10))
```

```python
df_covid_sum.loc[df_covid_sum.Country_Region=='US','Country_Region'] = 'United States of America'
```

```python
## left outer join
world_covid = pd.merge(world, df_covid_sum, left_on='name', right_on='Country_Region')
```

```python
world_covid.head()
```

```python
world_covid.plot(column='Confirmed', cmap='OrRd', scheme='percentiles', figsize=(20,10))
```
